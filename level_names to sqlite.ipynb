```python
import pymongo

client = pymongo.MongoClient("mongodb://localhost:27017/")
sefariaDB = client['sefaria']
```

```python
import pandas as pd
index = sefariaDB['index']
index = pd.DataFrame(list(index.find()))
index
```

```python
simple_books = index.loc[index['schema'].map(lambda x: True if 'sectionNames' in x else False)]
```

```python
simple_books['level_1_name'] = simple_books['schema'].apply(lambda x: x['sectionNames'][0] if 'sectionNames' in x else None)
simple_books['level_2_name'] = simple_books['schema'].apply(lambda x: x['sectionNames'][1] if 'sectionNames' in x and len(x['sectionNames'])>1 else None)
simple_books['level_3_name'] = simple_books['schema'].apply(lambda x: x['sectionNames'][2] if 'sectionNames' in x and len(x['sectionNames'])>2 else None)
simple_books['level_4_name'] = simple_books['schema'].apply(lambda x: x['sectionNames'][3] if 'sectionNames' in x and len(x['sectionNames'])>3 else None)
simple_books['level_5_name'] = simple_books['schema'].apply(lambda x: x['sectionNames'][4] if 'sectionNames' in x and len(x['sectionNames'])>4 else None)
simple_books['level_6_name'] = simple_books['schema'].apply(lambda x: x['sectionNames'][5] if 'sectionNames' in x and len(x['sectionNames'])>5 else None)
```

```python
complex_books = index.loc[index['schema'].map(lambda x: True if 'nodes' in x else False)]
```

```python
complex_books.info()
```

```python
complex_books_flattern_1 = (
    complex_books
    .assign(schema=complex_books.schema.map(lambda d: d['nodes']))
    .explode("schema")
    .reset_index(drop=True)
)
```

```python
complex_books_flattern_1 
```

```python
complex_books_flattern_1['level_1_index'] = complex_books_flattern_1['schema'].map(lambda x:x['key'] if 'key' in x else None)
simple_complex_books = complex_books_flattern_1.loc[complex_books_flattern_1['schema'].map(lambda x: True if 'sectionNames' in x else False)]
simple_complex_books = simple_complex_books.loc[simple_complex_books['schema'].map(lambda x: True if 'nodes' not in x else False)]
simple_complex_books
```

```python
simple_complex_books['level_2_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][0])
simple_complex_books['level_3_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][1] if len(x['sectionNames'])>1 else None)
simple_complex_books['level_4_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][2] if len(x['sectionNames'])>2  else None)
simple_complex_books['level_5_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][3] if len(x['sectionNames'])>3  else None)
```

```python
simple_complex_books
```

```python
index = pd.concat([simple_complex_books,simple_books])
```

```python
complex_books = complex_books_flattern_2.loc[complex_books_flattern_2['schema'].map(lambda x: True if 'nodes' in x else False)]
#complex_books = complex_books.loc[complex_books['level_1_index'].map(lambda x: False if type(x) == str else True)]
complex_books_flattern_3 = (
    complex_books
    .assign(schema=complex_books.schema.map(lambda d: d['nodes']))
    .explode("schema")
    .reset_index(drop=True)
)
complex_books_flattern_3 
complex_books_flattern_3['level_3_index'] = complex_books_flattern_3['schema'].map(lambda x:x['key'] if 'key' in x else None)
simple_complex_books = complex_books_flattern_3.loc[complex_books_flattern_3['schema'].map(lambda x: True if 'nodes' not in x else False)]
simple_complex_books = simple_complex_books.loc[simple_complex_books['schema'].map(lambda x: True if 'sectionNames'  in x else False)]
simple_complex_books
simple_complex_books['level_4_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][0])
simple_complex_books['level_5_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][1] if len(x['sectionNames'])>1 else None)
simple_complex_books['level_6_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][2] if len(x['sectionNames'])>2  else None)
simple_complex_books
index = pd.concat([index,simple_complex_books])
```

```python
complex_books = complex_books_flattern_3.loc[complex_books_flattern_3['schema'].map(lambda x: True if 'nodes' in x else False)]
#complex_books = complex_books.loc[complex_books['level_1_index'].map(lambda x: False if type(x) == str else True)]
complex_books_flattern_4 = (
    complex_books
    .assign(schema=complex_books.schema.map(lambda d: d['nodes']))
    .explode("schema")
    .reset_index(drop=True)
)
complex_books_flattern_4 
complex_books_flattern_4['level_4_index'] = complex_books_flattern_4['schema'].map(lambda x:x['key'] if 'key' in x else None)
simple_complex_books = complex_books_flattern_4.loc[complex_books_flattern_4['schema'].map(lambda x: True if 'nodes' not in x else False)]
simple_complex_books = simple_complex_books.loc[simple_complex_books['schema'].map(lambda x: True if 'sectionNames'  in x else False)]
simple_complex_books
simple_complex_books['level_5_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][0])
simple_complex_books['level_6_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][1] if len(x['sectionNames'])>1 else None)
simple_complex_books
index = pd.concat([index,simple_complex_books])
```

```python
complex_books = complex_books_flattern_4.loc[complex_books_flattern_4['schema'].map(lambda x: True if 'nodes' in x else False)]
#complex_books = complex_books.loc[complex_books['level_1_index'].map(lambda x: False if type(x) == str else True)]
complex_books_flattern_5 = (
    complex_books
    .assign(schema=complex_books.schema.map(lambda d: d['nodes']))
    .explode("schema")
    .reset_index(drop=True)
)
complex_books_flattern_5 
complex_books_flattern_5['level_5_index'] = complex_books_flattern_5['schema'].map(lambda x:x['key'] if 'key' in x else None)
simple_complex_books = complex_books_flattern_5.loc[complex_books_flattern_5['schema'].map(lambda x: True if 'nodes' not in x else False)]
simple_complex_books = simple_complex_books.loc[simple_complex_books['schema'].map(lambda x: True if 'sectionNames'  in x else False)]
simple_complex_books
simple_complex_books['level_6_name']=simple_complex_books['schema'].map(lambda x:x['sectionNames'][0])
simple_complex_books
index = pd.concat([index,simple_complex_books])
```

```python
default_values = index['level_1_index'].map(lambda x: True if x=="default" else False)
index.loc[default_values,'level_1_index']=index.loc[default_values,'level_2_index']
index.loc[default_values,'level_2_index']=index.loc[default_values,'level_3_index']
index.loc[default_values,'level_3_index']=index.loc[default_values,'level_4_index']
index.loc[default_values,'level_4_index']=index.loc[default_values,'level_5_index']

index.loc[default_values,'level_1_name']=index.loc[default_values,'level_2_name']
index.loc[default_values,'level_2_name']=index.loc[default_values,'level_3_name']
index.loc[default_values,'level_3_name']=index.loc[default_values,'level_4_name']
index.loc[default_values,'level_4_name']=index.loc[default_values,'level_5_name']
index.loc[default_values,'level_5_name']=index.loc[default_values,'level_6_name']
```

```python
import sqlite3

conn = sqlite3.connect('sefaria.sqlite')
```

```python

index[['title','level_1_index','level_1_name','level_2_index','level_2_name',
                'level_3_index','level_3_name','level_4_index','level_4_name','level_5_name', 'level_6_name'               
              ]].to_sql('level_names', conn, if_exists='replace', index=False)
conn.close()
```
