```python
import pymongo

client = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
pymongo.has_c()
```

```python
sefariaDB = client['sefaria']
```

```python
collections = sefariaDB.list_collections()
for collection in collections:
    print(collection['name'])
```

```python
index = sefariaDB['index']
```

```python
print(index.count_documents({}))
```

```python
import pandas as pd
index_df = pd.DataFrame(index.find())
index_df
```

```python
texts = sefariaDB['texts']
texts_df = pd.DataFrame(texts.find())
texts_df
```

```python
heb_texts = texts_df.loc[texts_df['language']=='he']
```

```python
heb_texts.count()
```

```python
heb_texts.drop_duplicates(subset=['title'],inplace=True)
```

```python
df = pd.merge(heb_texts,index_df,on='title')
```
