```python
def to_talmud(section)->str:
    if section == None:
        return None
    section = int(section) + 1
    daf = section // 2
    if section > daf * 2:
        daf = "{}b".format(daf)
    else:
        daf = "{}a".format(daf)
    return daf
```

```python
talmud_books = df_flattern_6['schema'].map(lambda x: True if 'addressTypes' in x and x['addressTypes'][0]=='Talmud' else False)
df_flattern_6.loc[talmud_books,'level_1_index'] = df_flattern_6.loc[talmud_books,'level_1_index'].map(lambda x: to_talmud(x) if type(x)==int else None)
```
